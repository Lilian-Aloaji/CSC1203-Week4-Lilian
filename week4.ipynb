```python
# 1) Program to sum even positive numbers

total = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    if num % 2 == 0:
        total += num

print("Final sum =", total)
```

```python
# 2) Multiplication table

num = int(input("Enter a number: "))

for i in range(1, 13):

    product = num * i

    if product > 50:
        break

    if product % 3 == 0:
        continue

    print(num, "x", i, "=", product)
```

```python
# 3) Guessing game

import random

secret = random.randint(1, 20)
attempts = 0

while True:
    guess = int(input("Guess the number (1-20): "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1   # augmented assignment

    if guess == secret:
        print("Correct! Attempts =", attempts)
        break
```

```python
# 4) Count numbers divisible by 4

n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count =", count)
```

```python
# 5) Triangle pattern

for i in range(1, 6):

    for j in range(1, i + 1):

        if j == 3:
            continue

        if i * j > 6:
            break

        print(j, end=" ")

    print()
```
